```python
import gmsh 

gmsh.initialize()
gmsh.clear()
gmsh.model.add("loaded_mesh")

mesh_path = "mesh_block_pipe"
gmsh.open(mesh_path + ".msh")

for _ in range(2):
    gmsh.model.mesh.refine()

# Save mesh
# WARNING!!! It's important to specify Version 2 GMsh file, otherwise there would be an error with NGSolve
gmsh.option.setNumber("Mesh.MshFileVersion", 2.1)
gmsh.write(mesh_path + "_refined" + ".msh")

# GUI
#     gmsh.fltk.run()

gmsh.finalize()

```
