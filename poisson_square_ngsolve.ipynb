```python
from ngsolve import *
from mpi4py import MPI
from netgen.meshing import Mesh
from netgen.occ import unit_square
import meshio
from ngsolve import *
from ngsolve.webgui import Draw 
from netgen.meshing import Mesh as NGMesh
import netgen
from netgen.geom2d import SplineGeometry
from netgen.read_gmsh import ReadGmsh
```

```python
# WARNING!!! If there is an error, it is likely because the version of .msh file is above 2.
msh = ReadGmsh("meshes/square/mesh.msh")
msh = Mesh(msh)

print("ElementBoundary=", msh.GetBoundaries())

Draw(msh)
```

```python
fes = H1(msh, order=1, dirichlet="Left|Right")

# Trial and test functions
u = fes.TrialFunction()
v = fes.TestFunction()

# Right-hand side
g = sin(6*x)
```

```python
a = BilinearForm(fes, symmetric=True)
a += grad(u)*grad(v)*dx

f = LinearForm(fes)
f += 10 * g * v * ds("Top")
f += 10 * g * v * ds("Bottom")

# f += 32 * (y*(1-y)+x*(1-x)) * v * dx

pre = preconditioners.Local(a)
a.Assemble()
f.Assemble()

gfu = GridFunction(fes)
```

```python
inv = CGSolver(a.mat, pre.mat)
gfu.vec.data = inv*f.vec

Draw(gfu)
```
