```python
import numpy as np
import pyvista as pv

sol_ngsolve = pv.read("../output/interactive/interactive_ngsolve.vtu")
sol_ngsolve = sol_ngsolve["sol"]
sol_fenicsx = pv.read("../output/interactive/interactive_fenics.vtu")
sol_fenicsx = sol_fenicsx["sol"]

abs_error = np.abs(sol_ngsolve - sol_fenicsx)

abs_error_max = np.max(abs_error)
abs_error_mean = np.mean(abs_error)

print(f"Max. absolute error between NGSolve and FEniCS : {abs_error_max}.")
print(f"Avg. absolute error between NGSolve and FEniCS : {abs_error_mean}.")

mesh_path = "../meshes/mesh_square.msh"

res = pv.read(mesh_path)

res['absolute_error_ngsolve_fenics'] = abs_error

res.save("../output/interactive/error_ngsolve_fenics.vtu")
```
