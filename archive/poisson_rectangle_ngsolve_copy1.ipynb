```python
# Convert .mesh file to .msh using GMsh
# Convert .mesh or .msh files to .vol using netgen
# netgen -geofile=mesh.msh -meshfile=mesh.vol


filename="meshes/mesh.vol"
msh = Mesh(filename)

Draw(msh)
```

```python
fes = H1(msh, order=1, dirichlet=[1,2,3,4])

# Trial and test functions
u = fes.TrialFunction()
v = fes.TestFunction()

# Right-hand side
g = sin(5*x)
```

```python
a = BilinearForm(fes, symmetric=True)
a += grad(u)*grad(v)*dx

f = LinearForm(fes)
# f += g * v * ds("left")
# f += g * v * ds("right")

f += 32 * (y*(1-y)+x*(1-x)) * v * dx

pre = preconditioners.Local(a)
a.Assemble()
f.Assemble()

gfu = GridFunction(fes)
```

```python
inv = CGSolver(a.mat, pre.mat)
gfu.vec.data = inv*f.vec

Draw(gfu)
```
