```python
import numpy as np
import pyvista as pv

sol_ngsolve = pv.read("output/square/poisson_square_ngsolve.vtu")
sol_ngsolve = sol_ngsolve["sol"]
sol_fenicsx = pv.read("output/square/poisson_sqaure_fenics.vtu")
sol_fenicsx = sol_fenicsx["sol"]

rel_error = np.abs(sol_ngsolve - sol_fenicsx)

rel_error_max = np.max(rel_error)
rel_error_mean = np.mean(rel_error)

print(f"Max. absolute error between NGSolve and FEniCS : {rel_error_max}.")
print(f"Avg. absolute error between NGSolve and FEniCS : {rel_error_mean}.")
```
