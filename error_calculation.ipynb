```python
import numpy as np
import pyvista as pv

case_name = "case3"
mesh_name = "mesh_block_pipe_refined"

sol_ngsolve = pv.read(f"output/{case_name}/{case_name}_ngsolve.vtu")
sol_ngsolve = sol_ngsolve["sol"]
sol_fenicsx = pv.read(f"output/{case_name}/{case_name}_fenics.vtu")
sol_fenicsx = sol_fenicsx["sol"]
sol_moose = pv.read(f"output/{case_name}/{case_name}_moose.vtu")
sol_moose = sol_moose["sol"]

# NGSolve and FEniCS

abs_error = np.abs(sol_ngsolve - sol_fenicsx)

abs_error_max = np.max(abs_error)
abs_error_mean = np.mean(abs_error)

print (f"Case: {case_name}")
print("")
print(f"Max. absolute error between NGSolve and FEniCS : {abs_error_max:.3e}.")
print(f"Avg. absolute error between NGSolve and FEniCS : {abs_error_mean:.3e}.")

mesh_path = f"meshes/{mesh_name}.msh"

res = pv.read(mesh_path)

res['absolute_error_ngsolve_fenics'] = abs_error

res.save(f"output/{case_name}/{case_name}_error_ngsolve_fenics.vtu")

# NGSolve and MOOSE

mesh_path = f"meshes/{mesh_name}.msh"

abs_error = np.abs(sol_ngsolve - sol_moose)

abs_error_max = np.max(abs_error)
abs_error_mean = np.mean(abs_error)

print(f"Max. absolute error between NGSolve and MOOSE : {abs_error_max:.3e}.")
print(f"Avg. absolute error between NGSolve and MOOSE : {abs_error_mean:.3e}.")

res = pv.read(mesh_path)

res['absolute_error_ngsolve_fenics'] = abs_error

res.save(f"output/{case_name}/{case_name}_error_ngsolve_moose.vtu")
```
